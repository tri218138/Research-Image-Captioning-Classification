```python
from datasets import load_dataset
```

```python
dataset = load_dataset("TeeA/Pokemon-Captioning-Classification")
dataset
```

```python
labels = dataset['train'].features["labels"].names
label2id, id2label = dict(), dict()
for i, label in enumerate(labels):
    label2id[label] = i
    id2label[i] = label

id2label[2]
```

```python
import os
import shutil
```

```python
for idx, data in enumerate(dataset['train']):
    print(data['labels'])
    print(type(data['labels']))
    data['image'].save("alibaa.png")
    break
```

```python
for split in dataset.keys():
    for idx, data in enumerate(dataset[split]):
        folder_dest = f"dataset/{split[:min(len(split), 5)]}/{id2label[data['labels']]}"
        os.makedirs(folder_dest, exist_ok=True)
        data['image'].save(f"{folder_dest}/{id2label[data['labels']]}_{idx}.png")
        # break
```
